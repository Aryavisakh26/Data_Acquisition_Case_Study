## Q13. Identify any potential outliers in a numerical column of your choice. Explain your approach.


```python
df.head()
```

```python
# Write your code here
Q1 = df['temp'].quantile(0.25)
Q3 = df['temp'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['temp'] < lower_bound) | (df['temp'] > upper_bound)]

print(f"Number of outliers in 'temp': {outliers.shape[0]}")
```

## Q.16 In which season (Spring, Summer, Fall, Winter) people rented bikes the most?

```python
# Write your code here
season_map = {1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"}
season_rentals = df.groupby("season")["cnt"].sum().idxmax()
print(f"Most rentals happened in: {season_map[season_rentals]}")
```

## Q.17 What is the peak hour in which bike rents the most?

```python
# Write your code here
peak_hour = df.groupby("hr")["cnt"].sum().idxmax()
print(f"Peak hour for bike rentals: {peak_hour}:00")
```

## Q.18 In which day of a week bikes rents out most?

```python
# Write your code here
# 'weekday' column: 0=Sunday, 6=Saturday
day_rentals = df.groupby("weekday")["cnt"].sum().idxmax()
print(f"Most bikes rented on weekday number:{day_rentals}")
```

## Q.19 In which hour Casual users rents bikes the most?

```python
# Write your code here
peak_hour_casual = df.groupby("hr")["casual"].sum().idxmax()
print(f"Peak hour for casual users: {peak_hour_casual}:00")
```

## Q.20 What is the maximum temperature observed in each of the seasons?

```python
# Write your code here
max_temp_by_season = df.groupby("season")["temp"].max()
max_temp_by_season.index = max_temp_by_season.index.map(season_map)
print(max_temp_by_season)
```
